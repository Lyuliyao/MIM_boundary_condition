```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
loss_periodic_2d = np.load("loss_periodic_2d.npy")
error_periodic_2d = np.load("error_periodic_2d.npy")
loss_periodic_4d = np.load("loss_periodic_4d.npy")
error_periodic_4d = np.load("error_periodic_4d.npy")
loss_periodic_8d = np.load("loss_periodic_8d.npy")
error_periodic_8d = np.load("error_periodic_8d.npy")
loss_periodic_16d = np.load("loss_periodic_16d.npy")
error_periodic_16d = np.load("error_periodic_16d.npy")
# print(len(loss_periodic_2d))
# print(len(loss_periodic_4d))
# print(len(loss_periodic_8d))
# print(len(loss_periodic_16d))
```

```python
epoch_1 = [i for i in range(len(loss_periodic_2d))]
loss_2d = list(loss_periodic_2d)
error_2d = list(error_periodic_2d)
loss_4d = list(loss_periodic_4d)
error_4d = list(error_periodic_4d)
epoch_2 = [i for i in range(len(loss_periodic_8d))]
loss_8d = list(loss_periodic_8d)
error_8d = list(error_periodic_8d)
epoch_4 = [i for i in range(len(loss_periodic_16d))]
loss_16d = list(loss_periodic_16d)
error_16d = list(error_periodic_16d)
```

```python
print("error 2d is ", error_2d[-1])
print("error 4d is ", error_4d[-1])
print("error 8d is ", error_8d[-1])
print("error 16d is ", error_16d[-1])
```

```python
plt.figure(figsize=(12,10))
plt.plot(epoch_1, loss_2d, color = "r", linestyle= "-.", marker = ".", linewidth = 0.5, label = "loss 2D")
plt.plot(epoch_1, loss_4d, color = "k", linestyle= "-.", marker = ".", linewidth = 0.5, label = "loss 4D")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend(loc = 'upper left', bbox_to_anchor = (0.7, 0.95))
# plt.savefig("loss_1.eps", dpi = 120)
plt.show() 
```

```python
plt.figure(figsize=(12,10))
plt.plot(epoch_1, error_2d, color = "r", linestyle= "-.", marker = ".", linewidth = 0.5, label = "error 2D")
plt.plot(epoch_1, error_4d, color = "k", linestyle= "-.", marker = ".", linewidth = 0.5, label = "error 4D")
plt.xlabel("epoch")
plt.ylabel("error")
plt.legend(loc = 'upper left', bbox_to_anchor = (0.7, 0.95))
# plt.savefig("error_1.eps", dpi = 120)
plt.show() 
```

```python
plt.figure(figsize=(12,10))
plt.plot(epoch_2, loss_8d, color = "r", linestyle= "-.", marker = ".", linewidth = 0.5, label = "loss 8D")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend(loc = 'upper left', bbox_to_anchor = (0.7, 0.95))
# plt.savefig("loss_2.eps", dpi = 120)
plt.show() 
```
