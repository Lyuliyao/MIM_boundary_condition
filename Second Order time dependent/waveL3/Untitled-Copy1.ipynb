```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
```

```python
error1 = np.load("error_save_DGM_wave_2D_s_w10.npy")[0]
error2 = np.load("error_save_LDG_wave_2D_s_w10_v1.npy")[0]
error3 = np.load("error_save_LDG_wave_2D_s_w10.npy")[0]
index1 = np.arange(0,np.size(error1),np.size(error1)//100)
index2 = np.arange(0,np.size(error2),np.size(error2)//100)
index3 = np.arange(0,np.size(error3),np.size(error3)//100)

fig, ax = plt.subplots(1, 1, figsize=(5,5))
ax.semilogy(index1,error1[index1],label="DGM")
ax.semilogy(index2,error2[index2],label="MIM1")
ax.semilogy(index3,error3[index3],label="MIM2")


ax.legend(fontsize=15)
ax.set_xticks(np.arange(0,np.size(error2)+1,10000))
ax.tick_params(labelsize=20)
plt.ylabel('relative error',fontsize=20)
plt.xlabel('iterations',fontsize=20)
plt.subplots_adjust(left=0.3,right=0.9,top=0.85,bottom=0.25)
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 
ax.xaxis.set_major_formatter(formatter) 
plt.savefig(r"C:\\Users\\lyupi\\Documents\\boundary condition\\JCP\\figure\\Wave_2D_w10_l3_requ.eps")
```
