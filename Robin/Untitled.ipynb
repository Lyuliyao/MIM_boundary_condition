```python
import matplotlib.pyplot as plt
import numpy as np
import os
import matplotlib.ticker as ticker
```

```python
error1 = np.load("error_save_LDG_Robin_2D_w5.npy")[0]
error2 = np.load("error_save_LDG_Robin_4D_w10.npy")[0]
error3 = np.load("error_save_LDG_Robin_8D_w20.npy")[0]
error4 = np.load("error_save_LDG_Robin_16D_w40.npy")[0]
```

```python
index1 = np.arange(0,np.size(error1),np.size(error1)//100)
index = np.arange(0,np.size(error2),np.size(error2)//100)

fig, ax = plt.subplots(1, 1, figsize=(5,5))
ax.semilogy(index1,error1[index1],label="2D")
ax.semilogy(index,error2[index],label="4D")
ax.semilogy(index,error3[index],label="8D")
ax.semilogy(index,error4[index],label="16D")
ax.legend(fontsize=12)
ax.set_xticks(np.arange(0,np.size(error2)+1,20000))
ax.tick_params(labelsize=20)
plt.ylabel('relative error',fontsize=20)
plt.xlabel('iterations',fontsize=20)
plt.subplots_adjust(left=0.3,right=0.9,top=0.85,bottom=0.25)
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 
ax.xaxis.set_major_formatter(formatter) 
plt.savefig(r"C:\\Users\\lyupi\\Documents\\boundary condition\\JCP\\figure\\Robin1.eps")
```
