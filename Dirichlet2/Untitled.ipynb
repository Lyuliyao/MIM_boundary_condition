```python
import matplotlib.pyplot as plt
import numpy as np
import os
import matplotlib.ticker as ticker
os.getcwd()
os.chdir("C:\\Users\\lyupi\\Documents\\boundary condition\\code\\Dirichlet2\\") 
```

```python
error1 = np.load("error_save_LDG_La_2D_w10.npy")[0]
error2 = np.load("error_save_LDG_La_4D_w15.npy")[0]
error3 = np.load("error_save_LDG_La_8D_w20.npy")[0]
error4 = np.load("error_save_LDG_La_16D_w20.npy")[0]
error5 = np.load("error_save_LDG_La_32D_w35.npy")[0]
error6 = np.load("error_save_LDG_La_64D_w70.npy")[0]
error7 = np.load("error_save_LDG_La_128D_w144.npy")[0]
error8 = np.load("error_save_LDG_La_256D_w280.npy")[0]
```

```python
index1 = np.arange(0,np.size(error1),np.size(error1)//100)
index = np.arange(0,np.size(error2),np.size(error2)//100)

fig, ax = plt.subplots(1, 1, figsize=(5,5))
ax.semilogy(index1,error1[index1],label="2D")
ax.semilogy(index,error2[index],label="4D")
ax.semilogy(index,error3[index],label="8D")
ax.semilogy(index,error4[index],label="16D")


ax.legend(fontsize=15)
ax.set_xticks(np.arange(0,np.size(error2)+1,5000))
ax.tick_params(labelsize=20)
plt.ylabel('relative error',fontsize=20)
plt.xlabel('iterations',fontsize=20)
plt.subplots_adjust(left=0.3,right=0.9,top=0.85,bottom=0.25)
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 
ax.xaxis.set_major_formatter(formatter) 
plt.savefig(r"C:\\Users\\lyupi\\Documents\\boundary condition\\JCP\\figure\\Dirichlet1_1.eps")
```
