```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
df1 = np.load('MIM_error_parabolic_12d.npy')
df2 = np.load('MIM2_error_parabolic_12d.npy')
df3 = np.load('DGM_error_parabolic_12d.npy')
```

```python
index1 = np.arange(0,np.size(df1),np.size(df1)//100)
index2 = np.arange(0,np.size(df2),np.size(df2)//100)
index3 = np.arange(0,np.size(df3),np.size(df3)//100)

fig, ax = plt.subplots(1, 1, figsize=(5,5))
ax.semilogy(index1,df1[index1],label="MIM1")
ax.semilogy(index2,df2[index2],label="MIM2")
ax.semilogy(index3,df3[index3],label="DGM")


ax.legend(fontsize=15)
ax.set_xticks(np.arange(0,np.size(df1)+1,50000))
ax.tick_params(labelsize=20)
plt.ylabel('relative error',fontsize=20)
plt.xlabel('iterations',fontsize=20)
plt.subplots_adjust(left=0.3,right=0.9,top=0.85,bottom=0.25)
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 
ax.xaxis.set_major_formatter(formatter) 
plt.savefig(r"C:\\Users\\lyupi\\Documents\\boundary condition\\JCP\\figure\\parabolic_error_12D.eps")
```
